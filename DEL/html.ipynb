```python
from langchain_community.document_loaders import AsyncHtmlLoader

urls = ["https://www.scouts.org.uk/por/10-uniform-badges-and-emblems/"]
loader = AsyncHtmlLoader(urls)
docs = loader.load()
```

```python
from langchain_community.document_transformers import Html2TextTransformer

html2text = Html2TextTransformer()
docs_transformed = html2text.transform_documents(docs)
```

```python
print(docs_transformed[0].model_dump_json(indent=3))
```

```python
print(docs_transformed[0].page_content)
```

```python
content = docs_transformed[0].page_content

START = "# "
END = "****\n"

parsed = content[content.index(START):content.index(END)]

display(Markdown(parsed))
```
